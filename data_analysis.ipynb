```python
from data_cleaning import get_cleaned_dataset

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = get_cleaned_dataset()
```

# Análisis

La variable que creo que puede ser más diciente es la edad. Entonces empezamos graficándola tal cuál como vienen los datos. Se puede notar una gran presencialos de datos nulos, representados por la edad -1. Esto puede afectar analisis posteriores, por lo que se eliminarán del dataset.

```python
sns.set(style="whitegrid")

plt.figure(figsize=(7, 5))
sns.histplot(data['age'].dropna(), bins=30, kde=True)
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
data = data[data['age'] > 0]
```

Repitiendo la gráfica, ahora no tenemos datos nulos

```python
# 1. Distribución de la edad
sns.set(style="whitegrid")

plt.figure(figsize=(7, 5))
sns.histplot(data['age'].dropna(), bins=30, kde=True)
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

Por cada país la edad varía, siendo Chile y Argentina los países con mayor edad promedio, mientras que Uruguay tiene la menor edad promedio.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='country', y='age')
plt.title('Distribución de Edad por País')
plt.xlabel('País')
plt.ylabel('Edad')
plt.xticks(rotation=45)
plt.show()
```

La cantidad de personas según el ranking por país no es un comportamiento homogéneo, varía en cada uno, siendo en unos países predominante el ranking mid, en otros el lower y en otros el upper

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='country', hue='ranking', order=data['country'].value_counts().index)
plt.title('Distribución de Ranking por País')
plt.xlabel('País')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.legend(title='Ranking')
plt.show()
```

Pero al graficar la distribución de la edad por ranking y por país sí se puede notar una tendencia de que el promedio de edad más bajo corresponde al ranking lower, el ranking mid está en la mitad, y el rango upper tiene en promedio la mayor edad, esta tendencia se conserva más o menos entre los diferentes países. 

Esto me invita a plantear como hipótesis que es posible que el ranking esté relacionado con alguna variable socioeconómica, dado que a mayor edad, más personas pertenecen al ranking. Un ejemplo de una variable con un comportamiento similar es la riqueza, que suele ser mayor mientras más edad tenga la persona.

```python
plt.figure(figsize=(15, 10))
sns.violinplot(data=data, x='ranking', y='age', hue='country', split=True)
plt.title('Distribución de Edad por Ranking y País')
plt.xlabel('Ranking')
plt.ylabel('Edad')
plt.legend(title='País')
plt.show()
```
